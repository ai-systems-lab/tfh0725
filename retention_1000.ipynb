```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
files = '10NIIIS4/retention_2000'
```

```python
all_data = []
for fpath in os.listdir(files):
    try:
        df = pd.read_csv(os.path.join(files, fpath), delimiter=';')
        all_data.append(df.res.to_list())
    except Exception:
        print(fpath)
plt.plot(np.array(all_data).T, 'o', markersize=0.2)
plt.ylabel('-')
plt.xlabel('')
plt.grid(True, linestyle='--', linewidth=0.5)
```

```python
plt.hist(np.array(all_data).flatten(), bins=20)
plt.ylabel('-')
plt.xlabel('')
plt.grid(True, linestyle='--', linewidth=0.5)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
files = '10NIIIS4/park_2000_split_2'
```

```python
all_data = []
for fpath in os.listdir(files):
    try:
        df = pd.read_csv(os.path.join(files, fpath), delimiter=';')
        all_data.append(df.res.to_list()[-1])
    except AttributeError:
        print(fpath)
plt.hist(all_data, bins=20)
plt.ylabel('Кол-во')
plt.xlabel('Сопротивление')
plt.grid(True, linestyle='--', linewidth=0.5)
```
